# Axion-photon coupling limits vs axion mass

Axion mass-PQ scale relation:
\begin{equation}
m_{a}=5.70(7) \mu \mathrm{eV}\left(\frac{10^{12} \,\mathrm{GeV}}{f_{a}}\right)
\end{equation}

Axion-photon coupling
\begin{equation}
g_{a \gamma} \equiv \frac{\alpha}{2 \pi} \frac{C_{a\gamma}}{f_{a}}=2.0 \times 10^{-16} C_{a \gamma} \frac{m_{a}}{\mu \mathrm{eV}} \mathrm{GeV}^{-1}
\end{equation}

Model dependent constant:
\begin{equation}
C_{a\gamma} \equiv \frac{E}{N} - 1.92(4) = 
\begin{cases}
  -1.92, & {\rm KSVZ} \\
  0.75, & {\rm DFSZ\,I} \\
  -1.25, & {\rm DFSZ\,II}
\end{cases}
\end{equation}

```python
%matplotlib inline
from numpy import *
import matplotlib.pyplot as plt
from PlotFuncs import FigSetup, MySaveFig, BlackHoleSpins, FilledLimit, line_background

def PlotBound(ax,filename,edgecolor='k',facecolor='crimson',alpha=1,lw=1.5,y2=1e10,zorder=0.1,
              linestyle='-',skip=1,FillBetween=True,edgealpha=1,rescale_m=False,
              scale_x=1,scale_y=1,start_x=0,end_x=nan,MinorEdgeScale=1.5,AddMinorEdges=False,
              label='',label2='',fontsize=15,rotation=0,text_pos=[1,1],text_pos2=[1,1],textcolor='w',path_effects=line_background(1.5,'k')):
    dat = loadtxt(filename)
    if end_x/end_x==1:
        dat = dat[start_x:end_x,:]
    else:
        dat = dat[start_x:,:]
    dat[:,0] *= scale_x
    dat[:,1] *= scale_y
    if rescale_m:
        dat[:,1] = dat[:,1]/dat[:,0]
    if FillBetween:
        ax.fill_between(dat[0::skip,0],dat[0::skip,1],y2=y2,color=facecolor,alpha=alpha,zorder=zorder,lw=0)
    else:        
        ax.fill(dat[0::skip,0],dat[0::skip,1],color=facecolor,alpha=alpha,zorder=zorder,lw=0)
    ax.plot(dat[0::skip,0],dat[0::skip,1],color=edgecolor,zorder=zorder,lw=lw,linestyle=linestyle,alpha=edgealpha)
    if skip>1:
        ax.plot([dat[-2,0],dat[-1,0]],[dat[-2,1],dat[-1,1]],color=edgecolor,zorder=zorder,lw=lw,linestyle=linestyle,alpha=edgealpha)
    if AddMinorEdges:
        ax.plot([dat[-1,0],dat[-1,0]],[dat[-1,1],MinorEdgeScale*dat[-1,1]],color=edgecolor,zorder=zorder,lw=lw,linestyle=linestyle,alpha=edgealpha)
        ax.plot([dat[0,0],dat[0,0]],[dat[0,1],MinorEdgeScale*dat[0,1]],color=edgecolor,zorder=zorder,lw=lw,linestyle=linestyle,alpha=edgealpha)

    if label!='':
        ax.text(text_pos[0],text_pos[1],label,fontsize=fontsize,rotation=rotation,color=textcolor,path_effects=path_effects,clip_on=True)
    if label2!='':
        ax.text(text_pos2[0],text_pos2[1],label2,fontsize=fontsize,rotation=rotation,color=textcolor,path_effects=path_effects,clip_on=True)
    return

fig,ax = FigSetup(Shape='Rectangular',ylab='$|g_{a\gamma\gamma}|$ [GeV$^{-1}$]',mathpazo=True)


# Helioscopes
PlotBound(ax,"limit_data/AxionPhoton/CAST_highm.txt",facecolor=[0.5, 0.0, 0.13],zorder=0.14,label=r'{\bf CAST}',fontsize=29,text_pos=[1e-2,9e-10])
PlotBound(ax,"limit_data/AxionPhoton/CAST.txt",facecolor=[0.5, 0.0, 0.13],zorder=0.14)

# Stellar cooling/Supernovae
PlotBound(ax,"limit_data/AxionPhoton/SolarNu.txt",facecolor='seagreen',zorder=0.06,text_pos=[1e1,2e-9],fontsize=33,label=r'{\bf Solar $\nu$}')
PlotBound(ax,"limit_data/AxionPhoton/GlobularClusters.txt",facecolor=[0.0, 0.66, 0.42],zorder=0.05,text_pos=[1e0,1.1e-10],fontsize=25,label=r'{\bf Globular clusters}')
PlotBound(ax,"limit_data/AxionPhoton/WhiteDwarfs.txt",facecolor='#2ec763',zorder=0.04,text_pos=[0.5e6,8e-8],fontsize=16,label=r'\noindent {\bf White}\newline  {\bf dwarfs}',rotation=87,path_effects=line_background(1,'k'))
PlotBound(ax,"limit_data/AxionPhoton/DiffuseGammaRays.txt",facecolor='#318c49',zorder=0.0299,text_pos=[1.5e5,1.5e-10],fontsize=18,label=r'{\bf Diffuse}-$\gamma$',path_effects=line_background(1,'k'))
#PlotBound(ax,"limit_data/AxionPhoton/SNe-decay.txt",facecolor='#15732e',zorder=0.03,text_pos=[4.5e7,0.3e-8],fontsize=18,label=r'{\bf Low-E SNe}')
PlotBound(ax,"limit_data/AxionPhoton/SN1987A_decay.txt",facecolor='#067034',zorder=0.029,text_pos=[1.5e5,0.95e-11],fontsize=15,label=r'{\bf SN1987A}-$\gamma$',rotation=-25.5,path_effects=line_background(1,'k'))
PlotBound(ax,"limit_data/AxionPhoton/SN1987A_HeavyALP_nu.txt",facecolor='darkgreen',zorder=0.03,text_pos=[6e5,8e-9],fontsize=15,label=r'{\bf SN87}-$\nu$',path_effects=line_background(1,'k'))

```
